Correlation to Heatmap

A correlation heatmap is a graphical representation of the correlation matrix, which shows the correlation coefficients between variables. Here's a step-by-step explanation:

Correlation Matrix: This is a table that shows the correlation coefficients between sets of variables. Each cell in the table shows the correlation between two variables. The value is between -1 and 1:

1 indicates a perfect positive correlation.
-1 indicates a perfect negative correlation.
0 indicates no correlation.
Heatmap: A heatmap is a data visualization technique that represents data in matrix form with colors. In the context of a correlation matrix:

Each cell's color represents the correlation coefficient between the variables.
Typically, colors range from dark blue (strong negative correlation) to dark red (strong positive correlation), with shades of lighter colors in between for weaker correlations.
Purpose: Correlation heatmaps help identify relationships between variables, making it easier to see which variables are positively or negatively correlated and to what extent.

```python
# Create heatmap
sns.heatmap(df.corr())
plt.title('Correlation Heatmap')
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Univariate Analysis

A joint plot is a type of visualization provided by the Seaborn library in Python, which allows you to analyze the relationship between two variables. It combines a scatter plot (or another type of bivariate plot) with univariate plots (such as histograms or kernel density estimates) on the margins. This combination helps you understand both the individual distributions of each variable and their combined distribution.

Key Features of a Joint Plot

Scatter Plot: Shows the relationship between two variables.
Marginal Plots: Display the distribution of each variable independently.
KDE Plot: You can also use kernel density estimates to show the density of the data points.
Customization: Various options to customize the appearance and types of plots.

PAIR PLOT

 pair plot (or pairplot) is a multi-plot grid provided by the Seaborn library in Python, 
which creates a matrix of scatter plots (or other types of bivariate plots) for each pair of variables in a dataset, 
along with univariate plots (such as histograms or kernel density estimates) on the diagonal. This visualization helps in understanding
the relationships between multiple variables and examining the distribution of each variable.

```python
sns.pairplot(df)
```

Distplot

distplot (distribution plot) is a function in the seaborn library in Python that combines a histogram with a
kernel density estimate (KDE) of the data. This provides a visual representation of the data distribution, which can help in understanding the 
data's underlying distribution pattern, density, and spread. The distplot function is often used for exploratory data analysis.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]

# Create a distplot
sns.distplot(data, hist=True, kde=True, 
             bins=5, color='blue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 2})

# Show the plot
plt.show()

plt.show()
```

A violin plot is a method of plotting numeric data and is similar to a box plot with a rotated kernel density plot on each side. It provides a visual summary of the data's distribution and its probability density. Here are the key components of a violin plot:

Kernel Density Estimation (KDE): The width of the plot at different y-values indicates the density of the data at those values. The shape of the violin plot is determined by the KDE.
Median and Quartiles: Often, a violin plot includes a marker for the median of the data and lines or markers to indicate the interquartile range (IQR).
Symmetry: Violin plots are typically symmetric, with the KDE mirrored on both sides of a central axis.
Comparison Across Categories: Multiple violin plots can be plotted side by side to compare distributions across different categories or groups.

Used to see both distribution of data in terms and KDE Kernel Density Estimation and Box plot

```python
# Violin Plot
sns.violinplot(x='size', y='total_bill',data=df)
plt.title('Violin Plot of Total Bill by tip')
plt.show()
```

```python
# Violin Plot
sns.violinplot(x='total_bill', y='day',data=df,palette='rainbow')
plt.title('Violin Plot of Total Bill by day')

# Inside is the box plot
# and curve graph is how data is distributed with respect to KDE
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')
iris.head(5)
```

```python
# # Set a theme
# sns.set_theme(style="whitegrid")

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x="species", y="sepal_length", data=iris, palette="Set3")
plt.title('Boxplot of Sepal Length by Species')
plt.show()
```

```python

# Countplot
plt.figure(figsize=(10, 6))
sns.countplot(x="species", data=iris, palette="Set2")
plt.title('Countplot of Species')
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x="species", y="sepal_length", data=iris, palette="muted", inner="quartile")
plt.title('Violin Plot of Sepal Length by Species')
plt.show()
```

```python

# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x="species", y="sepal_length", data=iris, palette="coolwarm", ci=None)
plt.title('Bar Plot of Sepal Length by Species')
```

```python
# Displot
plt.figure(figsize=(10, 6))
sns.displot(iris, x="sepal_length", hue="species", kind="kde", fill=True)
plt.title('Displot of Sepal Length by Species')
```

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(iris, x="sepal_length", hue="species", multiple="layer", palette="rainbow", kde=True)
plt.title('Histogram of Sepal Length by Species')
```
