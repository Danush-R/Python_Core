MatplotLib Tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

Generally easy to get started for simple plots Support for custom labels and texts Great control of every element in a figure High-quality output in many formats Very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

When you use import matplotlib.pyplot as plt and %matplotlib inline, you're setting up your Jupyter Notebook environment to visualize data using Matplotlib. Here's a brief explanation of each:

**import matplotlib.pyplot as plt:**

This imports the pyplot module from the matplotlib library, allowing you to create various types of plots and visualizations. You typically use plt as an alias to keep the code concise.


%matplotlib inline:
**bold text**
This is a magic command in Jupyter Notebooks. When you use %matplotlib inline, it configures the notebook to display Matplotlib plots directly within the notebook cells. This makes it convenient to see the visualizations immediately after running the cell without needing to save or open separate files.

```python
# compute the x and y coordinates for a sine curve
x=np.arange(0,4*np.pi,0.1) #stepsize - 0.1
y=np.sin(x)

plt.title('sine wave form')

#plot the points using matplotlib
plt.plot(x,y)
```

```python
# compute the x and y coordinates for a sine and cosine curve
x=np.arange(0,5*np.pi,0.1) #stepsize - 0.1
y_sin=np.sin(x)
y_cos= np.cos(x)

#subplot the points using matplotlib
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('sine wave form')

#subplot the points using matplotlib
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('cosine wave form')
```

```python
# Bar plot

x = [2,4,6,8,10] 
y = [11,13,15,17,19]  

x2 = [1,3,5,7,9] 
y2 = [12,14,16,18,20] 
plt.bar(x, y,color='blue') 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0.5, 0.2)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
